```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import  train_test_split
```

```python
dataset = pd.read_csv(r"C:\Users\sscomputers\Desktop\Daily 1 Project\Breast Cancer\breast cancer.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
# Raw feature
dataset.drop("Unnamed: 32", axis = 1, inplace = True)
```

```python
# The id feature isn't imp.
dataset.drop("id", axis = 1, inplace = True)
```

```python
dataset.head()
```

```python
# Encoding the taeget Value
dataset["diagnosis"] = dataset["diagnosis"].map({"M" : 1, "B" : 0})
```

```python
dataset.head()
```

```python
dataset["diagnosis"].value_counts()
```

```python
x = dataset.drop("diagnosis", axis = 1)
y = dataset["diagnosis"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 42)
```

```python
# Balacing the dataset

from imblearn.over_sampling import SMOTE
smote  = SMOTE(random_state = 42)

x_train, y_train = smote.fit_resample(x_train, y_train)
```

```python
#  Fitting the Model
from sklearn.linear_model import LogisticRegression

lg = LogisticRegression()
lg.fit(x_train, y_train)
```

```python
# Checking Accuracy
y_pred = lg.predict(x_test)

from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)*100
```

```python
# Prediction System

df = x_train.iloc[51]
np_df =  np.asarray(df)

prediction = lg.predict(np_df.reshape(1, -1))

if prediction[0] == 1:
    print("Cantrious")
else:
    print("Not Cantrious")
```
